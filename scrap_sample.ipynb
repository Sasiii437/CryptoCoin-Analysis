```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL to scrape
url = 'https://coinmarketcap.com/'

# Headers to act like a browser (optional, good practice)
headers = {'User-Agent': 'Mozilla/5.0'}

# Fetch the page
response = requests.get(url, headers=headers)

# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Check if we fetched content correctly
print(soup.title.text)
```

```python
# Extract all coin names based on the inspected class
names = soup.find_all('p', class_='sc-65e7f566-0 iPbTJf coin-item-name')

# Clean and collect text values
coin_names = [name.text.strip() for name in names]

# Convert to Pandas Series
coin_names_series = pd.Series(coin_names, name='Coin Name')

# Display
coin_names_series.head()
```

```python
# Extract all coin symbols
symbols = soup.find_all('p', class_='sc-65e7f566-0 byYAWx coin-item-symbol')

# Clean and collect text values
coin_symbols = [symbol.text.strip() for symbol in symbols]

# Convert to Series
coin_symbols_series = pd.Series(coin_symbols, name='Symbol')

# Display
coin_symbols_series.head()
len(coin_symbols_series)
```

```python
# Extract all price containers
prices = soup.find_all('div', class_='sc-142c02c-0 lmjbLF fall')

# Clean and collect text values
coin_prices = [price.find('span').text.strip() for price in prices]

# Convert to Series
coin_prices_series = pd.Series(coin_prices, name='Price (USD)')

# Display
coin_prices_series.head()
```

```python
# Extract 1h change values
changes_1h = soup.find_all('span', class_='sc-1e8091e1-0 jgYsZM')

# Clean and collect text values
coin_1h_change = [change.text.strip() for change in changes_1h]

# Convert to Series
coin_1h_change_series = pd.Series(coin_1h_change, name='1h Change (%)')

# Display
coin_1h_change_series.head()
```

```python
# Extract market cap values
market_caps = soup.find_all('span', class_='sc-11478e5d-1 jfwGHx')

# Clean and collect text values
coin_market_cap = [cap.text.strip() for cap in market_caps]

# Convert to Series
coin_market_cap_series = pd.Series(coin_market_cap, name='Market Cap')

# Display
coin_market_cap_series.head()
```

```python
# Extract volumes
volumes = soup.find_all('p', class_='sc-71024e3e-0 fOLOxZ font_weight_500')

# Clean and collect text values
coin_volumes = [volume.text.strip() for volume in volumes]

# Convert to Series
coin_volumes_series = pd.Series(coin_volumes, name='Volume')

# Display
coin_volumes_series.head()
```

```python
# Extract supply values
supplies = soup.find_all('div', class_='circulating-supply-value')

# Clean and collect text values
coin_supplies = [supply.find('span').text.strip() for supply in supplies]

# Convert to Series
coin_supply_series = pd.Series(coin_supplies, name='Circulating Supply')

# Display
coin_supply_series.head()
```
